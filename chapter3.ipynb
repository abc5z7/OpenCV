## 图像梯度计算

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

img = cv2.imread("src/img/lena.png")

def cv_show(img, name='img'):
    cv2.imshow(name, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.waitKey(1)
````

#### Sobel算子
```
     |-1 0 +1|            |-1 -2 -1|
Gx = |-2 0 +2| * A and Gy | 0  0  0| * A
     |-1 0 +1|            |+1 +2 +1|
``` 

````python
circle = cv2.imread("src/img/circle.JPG")
sobelx = cv2.Sobel(circle, cv2.CV_64F,1,0,ksize=3)

plt.imshow(sobelx)
````

````python
sobelx2 = cv2.Sobel(circle, cv2.CV_64F,1,0,ksize=3)
sobelx2 = cv2.convertScaleAbs(sobelx2)
cv_show(sobelx2)
````

````python
plt.imshow(sobelx2)
````

````python
sobelx3 = cv2.Sobel(circle, cv2.CV_64F,0,1,ksize=3)
sobelx3 = cv2.convertScaleAbs(sobelx3)
plt.imshow(sobelx3)
````

分别计算xy再相加

````python
sobelx = cv2.Sobel(circle, cv2.CV_64F,1,0,ksize=3)
sobelx = cv2.convertScaleAbs(sobelx)

sobely = cv2.Sobel(circle,cv2.CV_64F,0, 1,ksize=3)
sobely = cv2.convertScaleAbs(sobely)

sobelxy = cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)
plt.imshow(sobelxy)
````

不建议直接算，效果可能不好

````python
sobelxy2 = cv2.Sobel(circle, cv2.CV_64F,1, 1,ksize=3)
sobelxy2 = cv2.convertScaleAbs(sobelxy2)
plt.imshow(sobelxy2)
````

### Scharr算子-对图像更敏感
```
     |-3  0 +3 |            |-3 -10 -3|
Gx = |-10 0 +10| * A and Gy | 0   0  0| * A
     |-3  0 +3 |            |+3 +10 +3|
``` 

### laplacian算子

```
    |0  1  0|
G = |1 -4  1|
    |0  1  0|
```

````python
img = cv2.imread('src/img/lena.png',cv2.IMREAD_GRAYSCALE)
lap = cv2.Laplacian(img, cv2.CV_64F)
lap = cv2.convertScaleAbs(lap)

plt.imshow(lap)
````
